```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import RepeatedKFold
from numpy import mean
from numpy import std
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
data_df = pd.read_csv('service_length_double_precision.csv')
data_df = data_df.dropna(subset=['complaint_percentile', 'civilian_allegation_percentile', 'internal_allegation_percentile', 'trr_percentile', 'honorable_mention_percentile'])
data_df['complaint_percentile']
data_x = data_df[['civilian_allegation_percentile', 'internal_allegation_percentile', 'trr_percentile', 'complaint_percentile', 'allegation_count', 'sustained_count', 'civilian_compliment_count', 'discipline_count', 'trr_count', 'unsustained_count', 'honorable_mention_percentile']]
data_x_1 = data_df[['civilian_allegation_percentile', 'complaint_percentile', 'allegation_count', 'honorable_mention_percentile', 'major_award_count', 'sustained_count']]
data_y = data_df[['service_length']]
data_x
```

```python
model_reg = lgb.LGBMRegressor(learning_rate=0.06, num_leaves=100, max_depth=0)
model_reg_1 = lgb.LGBMRegressor(learning_rate=0.06, num_leaves=100, max_depth=0)
cv = RepeatedKFold(n_splits=10, n_repeats=5, random_state=1)
n_scores = cross_val_score(model_reg, data_x_1, data_y, scoring='r2', cv=cv, n_jobs=-1, error_score='raise')
n_scores_1 = cross_val_score(model_reg_1, data_x_1, data_y, scoring='r2', cv=cv, n_jobs=-1, error_score='raise')
print('full selected features MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
print('optimzed selected features MAE: %.3f (%.3f)' % (mean(n_scores_1), std(n_scores_1)))
```
